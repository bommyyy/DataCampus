## 고냉지배추(가격)

```python
# 품종이 '고냉지 배추'인 것만 추출
df_고냉지배추_price = df_배추[df_배추['품종']=='고냉지배추']
```

```python
df_고냉지배추_price.head()
```

```python
df_고냉지배추_price.info()
```

### 도매가격(원/kg) object->float 타입변환

```python
df_고냉지배추_price['도매가격(원/kg)']=pd.to_numeric(df_고냉지배추_price['도매가격(원/kg)'],errors='coerce')
```

```python
df_고냉지배추_price.info()
```

### 일별 도매가격(원/kg) 평균 구하기

```python
# 시계열 데이터로 만들기 위해 일별 도매가격(원/kg)평균 구하기
df_고냉지배추_price['일별 도매가격(원/kg) 평균']= df_고냉지배추_price.groupby(['일자'])['도매가격(원/kg)'].transform('mean')
```

```python
df_고냉지배추_price.head()
```

### 일별 거래금액(백만원) 평균 구하기

```python
# 시계열 데이터로 만들기 위해  일별 거래금액(백만원) 평균 구하기
df_고냉지배추_price['일별 거래금액(백만원) 평균']= df_고냉지배추_price.groupby(['일자'])['거래금액(백만원)'].transform('mean')
```

```python
df_고냉지배추_price
```

### 일별 거래량(톤) 합 구하기

```python
# 시계열 데이터로 만들기 위해 일별 거래량(톤) 합 구하기
df_고냉지배추_price['일별 거래량(톤) 합']= df_고냉지배추_price.groupby(['일자'])['거래량(톤)'].transform('sum')
```

```python
df_고냉지배추_price.head()
```

### 일자를 연도, 월, 일 로 분리

```python
df_고냉지배추_price['연도'] = df_고냉지배추_price['일자'].str.split('.').str[0]
df_고냉지배추_price['월'] = df_고냉지배추_price['일자'].str.split('.').str[1]
df_고냉지배추_price['일'] = df_고냉지배추_price['일자'].str.split('.').str[2]
```

```python
df_고냉지배추_price.head()
```

```python
df_고냉지배추_price.info()
```

```python
# 연도의 데이터 타입 변환
df_고냉지배추_price = df_고냉지배추_price.astype({'연도': 'int'})
```

```python
df_고냉지배추_price['연도'] = df_고냉지배추_price['연도']+2000
```

### 2달간 평균 데이터 추출 (%배추의 생육기간 : 2달)

```python
## 2달간 평균 데이터 추출

a=[]
e=[]

for i in range(128,2460):
    a.append(df_날씨_강릉.loc[i:i+60])

# print(e)

b=[]
for i in range(len(a)):
    b.append(a[i].mean())
    
# print(b)

    
c=pd.DataFrame(b[0]).T

for i in range(1,len(b)):
    d=pd.DataFrame(b[i]).T
    c = pd.concat([c,d])

c

c.to_csv('2달간 평균기후 데이터.csv',encoding='cp949')
# 2달간 평균기후 데이터.csv 저장 후 엑셀에서 '일시' 열 추가함.(2014-07-08부터 2020-11-24)
```
